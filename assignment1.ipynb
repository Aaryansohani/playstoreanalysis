```python
df["Size Bucket"]= pd.qcut(df["Size"],q=5).copy() #5 buckets (20% records in each) based on Size 
df
```

## By Content Rating vs. Size buckets, get the rating (20th percentile) for each combination 

```python
group=df.groupby(by=["Content Rating","Size Bucket"])["Rating"].quantile(.2).unstack() 
group
```

```python
sns.heatmap(group,annot=True) #Annotated heatmap
```

```python
ax = sns.heatmap(group,cmap='Greens',annot=True) #Greens color map 
```
